```python
for i,row in customers.iterrows():
    out = "\nAllCustomers.Add(new AppUser\n{"
    unique = str(int(row['Customer #']))
    out +="\n\t" + "UniqueIdentifier = "+ unique + ","
    out += "\n\tLastName = \"%s\"," %(row["Last Name"])
    out += "\n\tFirstName = \"%s\"," %(row["First Name"])
    out += "\n\tMiddleInitial = \"%s\"," %(row["Middle Initial"])
    date = row['Birthday']
    out += "\n\tBirthday = new DateTime(%s, %s, %s),"%(str(date)[:4], str(date)[5:7], str(date)[8:10])
    out += "\n\tAddress = \"%s\"," %(row['Street'])
    out += "\n\tCity = \"%s\"," %(row['City'])
    out += "\n\tState = \"%s\"," %(row['State'])
    zipcode = str(int(row['Zip']))
    out += "\n\tZip = \"%s\"," %(zipcode)
    phone = str(int(row["Phone Number"]))
    out += "\n\tPhoneNumber = \"%s\"," %(phone)
    out += "\n\tEmail = \"%s\"," %(row['Email'])
    out += "\n\tUserName = \"%s\"," %(row['Email'])
    points = str(int(row['PopcornPoints']))
    out += "\n\tDefaultPopcornPoints = %s"%(points)
    out += "\n});"
    out += "\n\nTmpPass.Add(\"%s\");" %(row["Password"])
    print(out)
```
