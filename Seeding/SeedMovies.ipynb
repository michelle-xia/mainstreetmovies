```python
xlsx = pd.ExcelFile('movie.xlsx')
movies = pd.read_excel(xlsx, 'Movies')
customers = pd.read_excel(xlsx, 'Customers')
employees = pd.read_excel(xlsx, 'Employees')
```

```python
# Making the C# code for movies
for i,row in movies.iterrows():
  beg = "\nAllMovies.Add(new Movie\n{"
  beg +="\n\t" + "UniqueIdentifier = "+ str(row["ID"]) + ","
  beg += "\n\t" + "Title = \"" + str(row["Title"]) + "\","
  beg += "\n\t" + "Genre = db.Genres.FirstOrDefault(c => c.GenreName == \""+ str(row["Genres"]) + "\"),"
  overview = str(row["Overview"])
  overview = overview.replace("\'", "\\'")
  overview = overview.replace('\"', '\\"')
  beg += "\n\t" + "MovieOverview = \"" + overview + "\","
  date = str(row["Release Date"])
  #beg += "\n\tReleaseYear = \"" + date[0:4] + "\","
  beg += f"\n\tReleaseDate = new DateTime({date[0:4]},{date[5:7]},{date[8:10]}),"
  beg += "\n\t" + "Revenue = " + str(row["revenue"]) + ","
  beg += "\n\t" + "Runtime = " + str(row["Running Time (minutes)"]) + ","
  beg += "\n\t" + "Tagline = \"" + str(row["tagline"]) + "\","
  mpaa = str(row['MPAA Rating']).replace("-", "")
  beg += "\n\t" + "MPAARating = MPAARatings." + mpaa + ","
  actors = (list(filter(None, [actorString.strip() for actorString in str(row["Actors"]).split(",")])))
  #actors = ''.join(actors)
  #beg += "\n\t" + "Actors = " + actors + ","
  beg += "\n\t" + "Actors = \""
  for i in range(len(actors)):
    if i <len(actors) - 1:
        beg += actors[i] + ", "
    else:
        beg += actors[i] + "\""
  beg += "\n});"
  print(beg)

  
```
